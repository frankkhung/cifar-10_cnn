CIFAR-10 is a image data set that contains planes, cars, birds, cats, deers, dogs, frogs, boats, and trucks. Comparing to MNIST dataset, there are more colors and more noises. Thus, the difficulties for identifying CIFAR-10 images are high. 

```python
from keras.datasets import cifar10
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
(x_img_train, y_label_train), (x_img_test, y_label_test) = cifar10.load_data()
```

Shape of the images

```python
x_img_train.shape # the size of the color is 3 because of RGB colors
pixel = 32
types = 10
```

Now, we want to create a dictionary to define each 

```python
label_dict = {0: "airplane", 1: "automobile", 2: "bird", 3: "cat", 4: "deer", 5: "dog", 6: "frog", 7: "horse", 8: "ship", 9 : "truck"}
```

```python
def plot_images_labels_prediction(images, labels, prediction, idx, num = 10):
    fig = plt.gcf() # Get current figure
    fig.set_size_inches(12, 14)
    if num > 25: num = 25
    for i in range(0, num):
        ax = plt.subplot(5, 5, 1+i) # create index for the images
        ax.imshow(images[idx], cmap = 'binary')
        title = str(i) + ' , ' + label_dict[labels[i][0]]
        if len(prediction) > 0:
            title += '=>' + label_dict[prediction[i]]
        ax.set_title(title, fontsize = 10)
        ax.set_xticks([]) # Ticks are the markers denoting data points on axes.
        ax.set_yticks([])
        idx+=1
    plt.show()
```

```python
plot_images_labels_prediction(x_img_train, y_label_train, [], 0, 10)
```

Now, we need to normalize the image by dividing $16*16-1 = 255$ after chaning its type. By doing this, the model's accuracy will be higher. 

```python
x_train_norm = x_img_train.astype('float32')/255.0
x_test_norm = x_img_test.astype('float32')/255.0
```

Because there are 10 types of things that are in the dataset, we want to categorize their types

```python
y_label_train_OneHot = np_utils.to_categorical(y_label_train)
y_label_test_OneHot = np_utils.to_categorical(y_label_test)
```

### Create a sequential model. Later on, we only need to put the NN into the model

```python
model = Sequential()
```

### Create the first Convolutional Layer

- filters = 32: randomly produce 32 filter weight (filter weight: can tune it during the training process) 
- kernal_size = (3, 3): each filter is 3X3 size (the common choice is to keep it 3X3 or 5X5)
- padding = 'same': make the image size the same
- input_shape = (32, 32, 3): first and second element is the size of the picture which is 32X32. For the third element, it represents RGB colours


```python
model.add(Conv2D(filters = 32, kernel_size = (3,3), input_shape = (32, 32, 3), activation = 'relu', padding = 'same'))
```

### Add in Dropout to avoid overfitting

- add into Dropout, it means that during every training it will random drop 25% of the neurons to avoid overfitting. 

```python
model.add(Dropout(rate = 0.25))
```

### Add a Max-Pooling Layer

- add Max-pooling, which means it decreases the size of the pictures. The amount of the picture produced wouldn't change.

```python
model.add(MaxPooling2D(pool_size = (2, 2)))
```

### Add the Second Convolution Layer and Max-Pooling Layer

- Produce 64 images out of it

```python
model.add(Conv2D(filters = 64, kernel_size = (3, 3), activation = 'relu', padding = 'same'))
```

```python
model.add(Dropout(0.25))
```

- use max-pooling to decrease the sample

```python
model.add(MaxPooling2D(pool_size = (2, 2)))
```

### Start to build the Neural Network

1. Flatten: make is an one-dimensional layer
2. Hidden Layer
3. Output Layer

```python
model.add(Flatten())
model.add(Dropout(rate = 0.25)) # to avoid overfitting
```

```python
model.add(Dense(pixel*pixel, activation = 'relu'))
model.add(Dropout(rate = 0.25))
```

```python
model.add(Dense(types, activation = 'softmax'))
```

### Define how the model is going to be trained

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
train_history = model.fit(x_train_norm, y_label_train_OneHot, validation_split = 0.2, epochs = 10, batch_size = 128, verbose = 1)
```

```python
model.save_weights("SaveModel/cifarCnnModel.h5")
print("Saved model to disk")
```

### Build a show_train_history function

```python
def show_train_history(train_history, train, validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title('Train History')
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.legend(['train', 'validation'], loc = 'upper left')
    plt.show()
```

### Plot the accuracy

- As the end of of the training, 'acc' is higher than 'val_acc' (it means overfitting)

```python
show_train_history(train_history, 'acc', 'val_acc')
```

### Plot the loss

```python
show_train_history(train_history,'loss', 'val_loss')
```

### Find the model accuracy through testing set 

- score returns: Scalar test loss and metrics values 

```python
scores = model.evaluate(x_test_norm , y_label_test_OneHot, verbose = 0) #verbose is just showing you the progress bar etc. 
```

### Start Making Predictions

```python
prediction = model.predict_classes(x_test_norm)
```

```python
plot_images_labels_prediction(x_img_test, y_label_test, prediction, 0, 10)
```

### Find the probability of the predictions

```python
predicted_prob = model.predict(x_test_norm)
```

### Build the show_Predicted_Probability function

```python
def show_Predicted_Prob(y, prediction, x_img, Predicted_Prob, i):
    print('label:', label_dict[y[i][0]], 'predict', label_dict[prediction[i]])
    plt.figure(figsize = (2, 2))
    plt.imshow(np.reshape(x_img_test[i], (32, 32, 3)))
    plt.show()
    for j in range(10):
        print(label_dict[j]+ ' Probability: %1.9f' %(Predicted_Prob[i][j]))
```
